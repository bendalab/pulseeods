```python
import csv
import os
import shutil

with open('file_paths.csv', 'r', encoding='ISO-8859-1', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    
    for i,row in enumerate(spamreader):
        if i>0:
            new_filename = '%03d_%s_%s-%s-%s_%s_%s_pulsefish'%(i,row[1],row[2],row[3].zfill(2),row[4].zfill(2),row[5],row[6])
            if len(row[7])>0:
                new_filename = new_filename + '_%s_wavefish'%(row[7])
            if len(row[10])>0:
                new_filename = new_filename + '_eel'
            new_filename = new_filename + '.wav'

            if len(row[11])>0:
                try:
                    shutil.copy(row[0][:row[0].index('.WAV')+4], 'test/%s'%new_filename)
                except:
                    shutil.copy(row[0][:row[0].index('.wav')+4], 'test/%s'%new_filename)
            else:
                try:
                    shutil.copy(row[0][:row[0].index('.WAV')+4], 'validation/%s'%new_filename)
                except:
                    shutil.copy(row[0][:row[0].index('.wav')+4], 'validation/%s'%new_filename)
```
